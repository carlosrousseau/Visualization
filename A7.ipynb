```python
path = 'datasets/antropometria-dataset.csv'
df= pd.read_csv(path, 
                low_memory=False)
df.head()
```

```python
 df['sistol'].plot(kind='hist', figsize=(20, 10))
plt.title('Histogram of systol')
plt.ylabel('Frecuencia')
plt.xlabel('numero')
plt.show()

 df['diastol'].plot(kind='hist', figsize=(20, 10))
plt.title('Histogram of diastol')
plt.ylabel('Frecuencia')
plt.xlabel('numero')
plt.show()
```

```python
 df['peso'].plot(kind='hist', figsize=(20, 10))
plt.title('Histogram of Weight')
plt.ylabel('Frecuencia')
plt.xlabel('numero')
plt.show()

 df['cintura'].plot(kind='hist', figsize=(20, 10))
plt.title('Histograma of Waist')
plt.ylabel('Frecuencia')
plt.xlabel('numero')
plt.show()
```

```python
sns.lmplot(x='peso', 
           y='cintura',
           data=df,
           fit_reg=False,
           hue= 'sexo'
          )
plt.title('Weight vs Waist', fontsize = 17)

plt.show()
```
