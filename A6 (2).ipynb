# Description of the Data

In this dataset of "pokemon.csv" we have a list of all the diferent pokemons, their name, their type, their attack, defense, hp, their total stats and if they are legendary or not.

# Interesting Observations

Some interesting observations that I discovered are that if your pokemon is legendary then his stats will tend to be higher and that non-legendary pokemon are weaker. I will prove this in the following 2 graphs.

```python
sns.lmplot(x='Attack', 
           y='Total',
           data=df,
           fit_reg=False,
           hue= 'Legendary'
          )
plt.title('Stats according to Legendary status', fontsize = 17)

plt.show()
```

```python
sns.lmplot(x='Attack', 
           y='Defense',
           data=df,
           fit_reg=False,
           hue= 'Legendary'
           
          )
plt.title('Attack and Defense stats by Legendary type', fontsize = 17)

plt.show()
```

# 1. Out of the 50 first pokemon, which type has more attack stats? 

```python
sns.swarmplot(x='Type 1',y='Attack',data=df.head(50))
plt.show
```

We can see by this graph that in the first 50 pokemons, the bug type pokemon has the best attack stats

# Which pokemon generation has the best total stats?

```python
sns.stripplot(x='Generation',y='Total',data=df,)
plt.show
```

we can see by this graph that the 1st and 3rd generation have the best total stats.
