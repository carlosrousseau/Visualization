# Description of Data
In this dataset we have a list of all the fifa soccer players, their ratings, and their stats

# Critical Variables Identified

* `overall_rating`: the overall player rating out of 100
* `value_euro`: the player's value

# Correlation

```python
plt.figure(figsize=(12,8))
plt.scatter(df['overall_rating'],df['value_euro'])
plt.xlabel('Overall Rating')
plt.ylabel('Value Euro')
plt.title('Overall Rating vs Value Euro')
```

# Boxplot of Overall Rating

```python
plt.figure(figsize=(12,8))
plt.boxplot(df['overall_rating'])
```

# Swarmplot

```python
plt.figure(figsize=(12,8))
sns.swarmplot(df.sample(1000)['overall_rating'])
```

```python
plt.figure(figsize=(12,8))
sns.swarmplot(df.sample(1000)['value_euro'])
```

# Pie Chart

```python
for i in df['overall_rating'].index:
  var = df['overall_rating'][i]

  if var > 70:
    df['overall_rating'][i] = '70<'
  elif var >= 65 and var <= 70:
    df['overall_rating'][i] = '65 - 70'
  else:
    df['overall_rating'][i] = '<65'
```

```python
plt.figure(figsize=(12,8))
plt.pie(df['overall_rating'].value_counts(), labels = df['overall_rating'].unique(), autopct = '%1.1f%%', shadow = True, explode = (0, 0.1, 0))
plt.title('Overall rating pie chart')
```

```python
plt.figure(figsize=(12,8))
plt.hist(df['value_euro'])
plt.ylabel('Frequency')
plt.xlabel('Value Euro')
plt.title('Value euro histogram')
```
