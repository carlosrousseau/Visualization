```python
path = 'datasets/antropometria-dataset.csv'
df= pd.read_csv(path, 
                low_memory=False)
df.head()
```

```python
 df['peso'].plot(kind='hist', figsize=(20, 10))
plt.title('Histograma de Peso')
plt.ylabel('Frecuencia')
plt.xlabel('Peso (kg)')
plt.show()
```

```python
 df['cintura'].plot(kind='hist', figsize=(20, 10))
plt.title('Histograma de Cintura')
plt.ylabel('Frecuencia')
plt.xlabel('Cintura')
plt.show()
```

```python
plt.figure()
plt.boxplot(df['peso'], whis=(0, 100))
```

```python
sns.lmplot(x='peso', 
           y='cintura',
           data=df,
           fit_reg=False,
           hue= 'sexo'
          )
plt.title('peso y cintura por sexo', fontsize = 17)

plt.show()
```
