```python
myPivot = pd.pivot_table(df, index = ['STNAME', 'CTYNAME'], 
                         values = ['BIRTHS2010'], 
                         aggfunc= ['count'], 
                         columns = 'REGION', 
                         margins = True,)

myPivot
```
