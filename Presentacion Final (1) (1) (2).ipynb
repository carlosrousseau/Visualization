# Description of Dataset
In this dataset we have a list of all the fifa soccer players, their ratings, and their stats

# Variables of Interest

* `overall_rating`: the overall player rating out of 100
* `value_euro`: the player's value





# Hypothesis

In this dataset, I predict that I can prove that if the overall rating is higher than the value of the player will be more.

# Descriptive statistics of Overall_rating and Value_euro:

```python
df['overall_rating'].describe()
```

```python
df['value_euro'].describe()
```

# 1. Does the overall rating has a relation with the value of the player?

```python
plt.figure(figsize=(12,8))
plt.scatter(df['overall_rating'],df['value_euro'])
plt.xlabel('Overall Rating')
plt.ylabel('Value Euro')
plt.title('Overall Rating vs Value Euro')
```

# Conclusion of Scatterplot
We can see that there is not a linear relation but an exponential relation that shows that as the overall rating get's higher the value of the euro gets expontentially higher.

# 2. Where is the majority of player ratings? And is there any outliers?


```python
plt.figure(figsize=(12,8))
plt.boxplot(df['overall_rating'])
```

# Conclusion of Boxplot
We can see that there is outliers above the rating of 86 that shows that they are very rare and also that the majority of player ratings are between 65-70. 

# 3. How are all the player ratings distributed?

```python
plt.figure(figsize=(12,8))
sns.swarmplot(df.sample(1000)['overall_rating'])
```

# Conclusion of Swarmplot
We can see that the majority of player ratings are distributed in between 60 and 80 and thatthere is outliers above 90 rating.

# 4. What is an outlier in player values?

```python
plt.figure(figsize=(12,8))
sns.swarmplot(df.sample(1000)['value_euro'])
```

# Conclusion of Swarmplot
We can see that the outlier in player values is above 2*10^7 which is 20 million euros or more.

# 5. What percent of players is rated 70 or higher, 65 or less, or in between 65-70

```python
for i in df['overall_rating'].index:
  var = df['overall_rating'][i]

  if var > 70:
    df['overall_rating'][i] = '70<'
  elif var >= 65 and var <= 70:
    df['overall_rating'][i] = '65 - 70'
  else:
    df['overall_rating'][i] = '<65'
```

```python
plt.figure(figsize=(12,8))
plt.pie(df['overall_rating'].value_counts(), labels = df['overall_rating'].unique(), autopct = '%1.1f%%', shadow = True, explode = (0, 0.1, 0))
plt.title('Overall Rating Distribution', fontsize = 25)
```

# Conclusion of Pie Chart
We can see that approximately 40% of the players are rated 70 or higher, 26% are rated 65 or below and 34% lie between 65 and 70.

# 6. What is the average value of a player?

```python
plt.figure(figsize=(12,8))
plt.hist(df['value_euro'])
plt.ylabel('Frequency')
plt.xlabel('Value Euro')
plt.title('Value euro histogram')
```

# Conclusion of Histogram

We can see that the average value of a player lies between .1 * 10^7 which is approximately 1 million euros


# 7. Which player costs more, if he is rated 65 or less, 65-70, or 70 or higher?

```python
plt.figure(figsize=(12,8))
sns.swarmplot(x = 'overall_rating', y = 'value_euro', data = df.sample(1000))
```

# Conclusion of Swarmplot

We can see a relation with value and ratign here as the players that have 70 or more rating cost more thatn those who are 65-70, or 65 or less.
