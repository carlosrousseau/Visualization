```python
df = pd.read_csv('datasets/Cartwheeldata.csv')
df.head
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'CWDistance',
           data = df,
           fit_reg = False,
           hue = 'Gender'
          )
plt.show()
```

# 2. Make a distribution of a categorical variable 'Gender' using swarmplot.

```python
sns.swarmplot(x = 'Gender',
              y = 'CWDistance',
              data = df)
plt.show()
```

```python
sns.boxplot(data = df.loc[:,['Height', 'Wingspan', 'CWDistance', 'CompleteGroup', 'Score']
                         ]
           )
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(rotation = 45, fontsize = 17)
plt.show()
```

```python
sns.boxplot(data = df.loc[df['Gender']=='M',['Score']]
           )
           
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(rotation = 45, fontsize = 17)
plt.show()
```
